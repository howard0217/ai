```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx')#在x=1,y=0四個位置繪製(g)色x 標記
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--')#繪製(r)紅色-標記也就是線(趨勢線)
#取亂數
X=1+np.arange(30)/10
delta=np.random.uniform(low=-0.1,high=0.1,size=(30,))
#帶入公式
Y=0.3*X-0.3+delta
plt.plot(X,Y,'bo')#藍色的圓點
plt.plot('X')#設定顯示X文字
plt.plot('Y')#設定顯示Y文字
plt.show()#繪圖
```

```python
import pandas as pd
from sklearn import linear_model#繪圖函示庫
import matplotlib.pyplot as plt#繪圖函示庫
#read data
dataframe=pd.read_fwf('/brain_body.txt')#讀取txt的數據
x_values=dataframe[['Body']]#身體的數據
y_values=dataframe[['Brain']]#大腦的數據
#訓練
body_reg=linear_model.LinearRegression()#使用線性回歸模型
body_reg.fit(x_values,y_values)
pre=body_reg.predict(x_values)
#訓練預測
print(body_reg.predict(pd.DataFrame(data=[170])))
#圖形化
plt.scatter(x_values,y_values)
plt.plot(x_values,pre)
plt.show()
```

```python
import matplotlib.pyplot as plt#繪圖函示庫
import numpy as np #矩陣函示庫
from sklearn import datasets,linear_model#線性回歸函示庫
#取得糖尿病的數據
diabetes=datasets.load_diabetes()
#只取得第三個特徵值BMI
diabetes_X=diabetes.data[:,np.newaxis,2]
#切分特徵值BMI
diabetes_X_train=diabetes_X[:-20]#切割0到最後20比特徵給訓練用
diabetes_X_test=diabetes_X[-20:]#切割最後20比特徵給訓練用
#切分答案
diabetes_y_train=diabetes.target[:-20]#切割0到最後20比答案給訓練用
diabetes_y_test=diabetes.target[-20:]#切割最後20比答案給訓練用
#計算
regr=linear_model.LinearRegression()#建立線性回歸
regr.fit(diabetes_X_train,diabetes_y_train)#訓練
#係數
print('Coefficients: \n',regr.coef_)
#均方誤差
print("Mean squared error: %.2f"%np.mean((regr.predict(diabetes_X_test) - diabetes_y_test)**2))
#顯示方差分數:1是完美預測
print('Variance score: %.2f' %regr.score(diabetes_X_test,diabetes_y_test))
#繪圖
plt.scatter(diabetes_X_test,diabetes_y_test,color='black')#劃出測試黑點
plt.plot(diabetes_X_test,regr.predict(diabetes_X_test),color='blue',linewidth=3)#劃出測試預測的線性回歸答案
plt.xticks(())
plt.yticks(())
plt.show()#顯示
```
